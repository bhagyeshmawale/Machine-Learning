```python
## Frequency of each word
```

```python
# Tokens : words in each text 
```

```python
text.split()
```

```python
words = nltk.word_tokenize(txt) # creates a list of words
senten = nltk.sent_tokenize(txt)
```

```python
senten
```

```python
from string import punctuation
```

```python
words = [word for word in words if word not in punctuation and word not in eng_stopwords]
```

```python
import pandas as pd
fre = pd.Series(nltk.probability.FreqDist(words))
fre.sort_values(ascending=False)
```

```python
wc.generate_from_frequencies(fre)
plt.figure(figsize = (15,15))
plt.imshow(wc)
plt.show()
```

```python
fre = nltk.probability.FreqDist(words)
fre.most_common(10)
```

```python
from string import punctuation
```

```python
final_words = [word for word in words if word not in eng_stopwords + list(punctuation)]
```

```python
fre = nltk.probability.FreqDist(final_words)
fre.most_common(10)
```
