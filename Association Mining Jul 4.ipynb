```python
import pandas as pd
```

```python
data=pd.read_excel('Apriori.xlsx')
```

```python
data.columns=data.columns.str.strip().str.replace(' ','_')
```

```python
data.Bill_Number.nunique()
```

```python
basket=data.pivot_table(index='Bill_Number',
                       columns='Item_Desc',
                       values='Quantity',
                        aggfunc='sum',
                        fill_value=0)
basket.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
data.Item_Desc.value_counts()[:20].plot.bar()
plt.show()
```

```python
def encode(x):
    if x>=1: return 1
    elif x<=0: return 0
```

```python
basket_sets=basket.applymap(encode)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_item_sets=apriori(basket_sets,min_support=.0001,use_colnames=True)
```
